```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


bike_data = pd.read_excel('99Bikers_Raw_data.xlsx')

bike_data = bike_data.dropna()


plt.figure(figsize=(10, 6))
sns.histplot(bike_data['list_price'], bins=20, kde=True)
plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

correlation_matrix = bike_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()


summary_stats = bike_data.describe()


bike_data['Profit'] = bike_data['list_price'] - bike_data['standard_cost']


print("Summary Statistics:\n", summary_stats)
print("\nCorrelation Matrix:\n", correlation_matrix)
print("\nDataset:\n", bike_data)
```
